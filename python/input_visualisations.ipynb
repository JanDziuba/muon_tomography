```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
import re
```

```python
all_files = glob.glob(os.path.join("../data/training_data", "*.csv"))
all_files.sort(key=lambda f: int(re.sub('\D', '', f)))

print(all_files[0])
```

```python
first_file = pd.read_csv(all_files[0], header=None)
first_file.head()
```

```python
# add a column to first_file named 'energy_diff' being the difference between second to last and last column
first_file['energy_diff'] = first_file.iloc[:, -2] - first_file.iloc[:, -1]

print(first_file['energy_diff'][:10])
```

```python
# plot first_file['energy_diff']

plt.plot(first_file['energy_diff'][:1000])
```

```python
# plot histogram of first_file['energy_diff']

plt.hist(first_file['energy_diff'][:1000])
```

```python
# make the histogram more readable by setting the number of bins to 100

plt.hist(first_file['energy_diff'][:1000], bins=100)
plt.show()
```

```python
# print mean of energy_diff
print(first_file['energy_diff'].mean())
```

```python
# plot third to last column of first_file

plt.hist(first_file.iloc[:, -3][:1000], bins=100)
plt.show()
```

```python
# create new column in first_file named 'energy_diff_prop' = 'energy_diff' / third to last column

first_file['energy_diff_prop'] = first_file['energy_diff'] / first_file.iloc[:, -3]

# plot histogram of 'energy_diff_prop'

plt.hist(first_file['energy_diff_prop'][:1000], bins=100)
plt.show()
```

```python
# plot histogram of 'energy_entering' column

plt.hist(first_file.iloc[:, -4][:1000], bins=100)
plt.show()
```
