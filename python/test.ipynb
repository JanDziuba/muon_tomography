```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
import re
```

```python
out_middle = pd.read_csv("./out_middle.csv", header=None)
out_line_converged = pd.read_csv("./out_line_converged.csv", header=None)
out_line_avg = pd.read_csv("./out_line_avg.csv", header=None)
out_line_1_it = pd.read_csv("./out_line_1_it.csv", header=None)
```
